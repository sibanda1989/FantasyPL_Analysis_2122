```python
import requests
import pandas as pd
import numpy as np
```

```python
#fantasy premier league api url
url = 'https://fantasy.premierleague.com/api/bootstrap-static/'

#using the url to make a GET request from the API endpoint
r = requests.get(url)

#transforming the request into a JSON object
json =  r.json()
```

```python
#lets have a look at the json keys
json.keys()
```

```python
#main interest is in 3 keys, teams, element_type and elements.
#building dataframes from these keys

elements_df = pd.DataFrame(json['elements'])
element_types_df = pd.DataFrame(json['element_types'])
teams_df = pd.DataFrame(json['teams'])
```

```python
teams_df.head()
```

```python
element_types_df.head()
```

```python
elements_df.head()
```

```python
elements_df.columns
```

```python
#elements df contains player data so we will make a copy of it and return only some useful columns
player_df = elements_df[['first_name', 'second_name', 'team', 'element_type', 'selected_by_percent', 'now_cost', 'transfers_in', 'value_season', 'total_points', 'minutes', 'goals_scored', 'assists', 'clean_sheets', 'goals_conceded', 'bonus', 'points_per_game']]

player_df.head()
```

```python
element_types_df.head()
```

```python
#merging two dfs to return a specific column, (similar to left join in SQL, vlookup in excel)
player_df1 = pd.merge(player_df, element_types_df[['id','singular_name']], left_on='element_type', right_on='id', how='left')

player_df1.head()
```

```python
teams_df.head()
```

```python
player_df = pd.merge(player_df1, teams_df[['id', 'name']], left_on='team', right_on='id')
player_df.head()
```

```python
player_df.drop(['team', 'id_x', 'id_y'], axis=1, inplace=True)


player_df.head()
```

```python
player_df.rename(columns={"singular_name": "position", "name": "team"}, inplace=True)

player_df.head()
```

```python
player_df.shape
```

```python
player_df.info()
```

```python
#converting player value to float
player_df['value'] = player_df.value_season.astype(float)
```

```python
#dropping value_season column
player_df.drop(['value_season'], axis=1, inplace=True)
```

```python
#sorting players by value
player_df.sort_values('value', ascending=False).head(10)
```

We can see that defenders and goalkeepers present the best value (points to price ratio)

```python
#creating a pivot on value per position
pivot = player_df.pivot_table(index='position', values='value', aggfunc='mean').reset_index()

pivot.sort_values('value', ascending=False)
```

Defenders represent the highest value, followed by midfielders

```python
player_df['bonus'] = player_df['bonus'].astype(int)
```

```python
player_df.drop(['element_type'], axis=1, inplace=True)
```

```python
#to get a true reflection of positional value, lets drop players who have not played. 
#These will have 0 value, since value = points/price (and points are only awarded to those that will have played) 

player_df = player_df.loc[player_df['value'] > 0]
```

```python
pivot = player_df.pivot_table(index='position', values='value', aggfunc='mean').reset_index()

pivot.sort_values('value', ascending=False)
```

Now goalkeepers are back on top

```python
#lets see which team provides best value

team_pivot = player_df.pivot_table(index='team', values='value', aggfunc='mean').reset_index()

team_pivot.sort_values('value', ascending=False)
```

```python
#points per minute played, ppm

player_df['ppm'] = player_df['total_points'] / player_df['minutes']
```

```python
#saving the data to a csv flat file
player_df.to_csv('fpl_data.csv', index=False)
```
